```python
df = pd.read_csv('https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/Pandas/CalIt2.data', header = None)
df
```

```python
df.columns
```

```python
#Renombrar columnas

df = df.rename(columns={0: 'Flow ID',
                        1: 'Date',
                        2: 'Time',
                        3: 'Count'})
df
```

```python
#Seleccionar datos
# 1 Seleccionar todos los datos para la fecha 24 de julio de 2005 con un flow id=7.
date_filter = df['Date']=='07/24/05'
date_filter.head()
```

```python
filter_date = df['Date']=='07/24/05'
dfnew = df.loc[filter_date, :]
dfnew[['Date', 'Flow ID','Count']]
```

```python
dfnew['Date'].value_counts()
```

```python
filter_flow_id = dfnew['Flow ID']==7
dfnew1 = dfnew.loc[filter_flow_id, :]
dfnew1[['Date', 'Flow ID','Count']]
```

```python
dfnew1['Count'].value_counts()
```

```python
# Con dos datos seleccionados en el paso 1,
# ordenen el resultado en Count en orden descendiente y saquen las 10 primeras filas. Asignen la salida a una variable llamada top 10.
dfnew1=dfnew1.sort_values(by='Count', ascending=False)
```

```python
top10 = dfnew1.head(10)
top10
```

```python
#Para las 10 filas que aparecen arriba, utilice la función Apply de Pandas para restar el valor más bajo de los 10 de todos ellos
# y luego obtengan el valor promedio de los recuentos resultantes
```

```python
min_value = top10['Count'].nsmallest(10).min()
min_value
```

```python
def descontarmin(x) :
    return x - min_value
top10['Count'] = top10['Count'].apply(descontarmin)
top10
```

```python
average = top10['Count'].mean()
print("El valor promedio de los recuentos resultantes es:", average)
```

```python
#Seleccionar los datos en el mes de agosto 2005 con un flow id=7


df['Date'] = pd.to_datetime(df['Date'])
```

```python
filter_agosto = (df['Date'].dt.month==8)&(df['Date'].dt.year == 2005)
df3 = df.loc[filter_agosto, :]
df3[['Date', 'Flow ID']]
```

```python
df3['Date'].value_counts()
```

```python
filter_flow7 = df3['Flow ID']==7
df4 = df3.loc[filter_flow7, :]
df4[['Date', 'Flow ID']]
```

```python
df4['Flow ID'].value_counts()
```

```python
## Agrupar los datos en función de la fecha y obtener el recuento máximo por fecha
df4.groupby(['Date'])['Count'].sum().reset_index()

```
